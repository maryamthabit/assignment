```python
%%writefile home.py
import streamlit as st
import pandas as pd
import altair as alt
import plotly.express as px
if st.button('welcome to our page'):
    st.write(' hello')
else:
    st.write('Goodbye')
sreen = st.sidebar.selectbox("select the page:",["page1","page2","page3"])
market = pd.read_csv('marketplace.csv')
if sreen == "page1":
       st.title('Air Traffic Application')
       st.text('San Francisco International Airport Report on Monthly Passenger Traffic Statistics by Airline')
acti = st.selectbox("Select Operating Airline",market['sku_name'].unique())
st.write(acti)
plot_type=st.radio("select the plot type",['scatter','line'])
if plot_type =='scatter':
  pl = alt.Chart(market[market['sku_name']==acti]).mark_circle().encode(
    x = 'total_sold',
    y ='total_review',
    tooltip = ['total_sold','total_review']
).interactive()
else:
      pl = alt.Chart(market[market['sku_name']==acti]).mark_line().encode(
    x = 'total_sold',
    y ='total_review',
    tooltip = ['total_sold','total_review']
).interactive()
st.altair_chart(pl)
```

```python
%%writefile page2.py
import streamlit as st
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
import plotly.express as px
market = pd.read_csv('marketplace.csv')
st.title('Air Traffic Application')

number = st.number_input('what you Suggest the number of Passenger Count')
st.write('The current number is ', number)
pie_chart = px.pie(market, 
            title='number of Passenger with the GEO Region',
            values='price',
            names='GEO Region')
st.plotly_chart(pie_chart)
```
